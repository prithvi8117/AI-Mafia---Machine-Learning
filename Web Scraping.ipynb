### Considerations :
- Check site's **terms and conditions** - like amazon don't allow
- Are you going to the DOS Attacks - **Space out your requests** and make sure that you don't overload that site's server
- **Keep a check on your scrapers,** so that you don't have to rewrite the code again and again
- Clean up your data, **preprocessing data contract.** As the data you would be getting is raw, and not in the required manner
- Every page and situation is different, so **Configure your scraper** each and every time

### We will be needing three modules for a web scraper.
- **BeautifulSoup** : Allow you to scrape a website - download the entire website, give you the content
- **lxml parser** : Get and parse (convert) the entire website information as HTML (As it may be json, xml, html)
- **requests** : To make a request or call the website

Request or call the website server by requests module -----> Extracting information and scraping it from the website by BeautifulSoup module -----> parsing by lxml parser module -----> Finally you have your Data

### Scrapy to scrape daily news article's month and Year in studentnewsdaily.com

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from pandas import Series, DataFrame
```

```python
url = 'https://www.studentnewsdaily.com/archive/daily-news-article/'
```

```python
# 1. Call the API Server using requests module

result = requests.get(url)      #get - fetching from the website server
print(result)
```

```python
#It is giving everything as a string - raw data

print(type(result.content), end='\n\n')
print(result.content)
```

```python
# Making Use of BeautifulSoup and lxml to get the data/content in clean html format

soup = BeautifulSoup(result.content, 'lxml')    
print(type(soup), end='\n\n')
print(soup)
```

```python
content = soup.findAll('section')    # Extract and return all occurrence of section in the soup
print(type(content), end='\n\n')
print(content)
```

```python
len(content)
```

```python
content[1]
```

```python
# Extracting and returning the first occurrence of 'col-md-10 col-lg-10 col-xl-10' class in the content[1]

summary = content[1].find(class_='col-md-10 col-lg-10 col-xl-10')
print(summary)
```

```python
# Extracting and returning the all occurrence of <h4> tag in the summary and saving it as a list l

l = summary.findAll('h4')
print(l)
```

```python
# omitting the html tags and saving only the string part to a list series

series = [ele.get_text() for ele in l]
print(series)
```

### Scraping imdb.com to scrape images(image links) of Most Popular Movies

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url2 = 'https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm'
```

```python
result2 = requests.get(url2)
print(result2)
```

```python
soup2 = BeautifulSoup(result2.content, 'lxml')
print(type(soup2), end='\n\n')
print(soup2)
```

```python
contents = soup2.findAll(class_='posterColumn')
print('Total Film posters in this page :', len(contents), end='\n\n')
print(contents)
```

```python
contents[0].find('img')['src']
```

```python
i = 1
for content_ob in contents:
    print('( Image',i,') :', content_ob.find('img')['src'], end='\n\n')
    i +=1
```

### OR We can do it like this

```python
images = soup2.findAll('img')
print(len(images), end='\n\n')
print(images)
```

```python
i = 1
for image in images:
    print('( Image',i,') Movie :',image['alt'],':\n', image['src'], end='\n\n')
    i +=1
```

```python
# Saving the image links in a dictionary

d = dict()

for image in images:
    d[image['alt']] = image['src']

print(d)
```

```python
# Saving the image links with their names in a DataFrame

images_df = pd.DataFrame()
images_df['Movie Name'] = list(d.keys())
images_df['Image Link'] = list(d.values())
images_df
```

```python
# Saving the images_df in a csv file with name imdbMostPopularMoviesImages.csv

images_df.to_csv('F:/Study 2020-21/Python/2020+/Coding Club/ML/Data/mdbMostPopularMoviesImages.csv', index=False, header=True)
```
