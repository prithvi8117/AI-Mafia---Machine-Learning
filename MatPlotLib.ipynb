### Plotting 2 lines on same plot

```python
# line 1 points
x1 = [1,2,3]
y1 = [2,4,1]

# line 2 points
x2 = [1,2,3]
y2 = [4,1,3]
```

```python
plt.title('Yo Graph, Two lines on same graph!')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
# plotting the line 1 points 
plt.plot(x1, y1, label = "line 1")

# plotting the line 2 points 
plt.plot(x2, y2, label = "line 2")
plt.legend()    # show labels for different lines in a legend (here at the upper right corner of the plot)
plt.show()
plt.close()
```

## Subplots

Subplots are required when we want to show two or more plots in same figure. We can do it in two ways using two slightly different methods.

### Plotting subplots using method 1

```python
# function to generate coordinates

def create_plot(ptype):
    
    # setting the x-axis vaues
    x = np.arange(-10, 10, 0.01)
     
    # setting the y-axis values
    if ptype == 'linear':
        y = x
    elif ptype == 'quadratic':
        y = x**2
    elif ptype == 'cubic':
        y = x**3
    elif ptype == 'quartic':
        y = x**4
             
    return(x, y)
```

**Figure acts as a top level container for all plot elements.**

```python
# create a figure

fig = plt.figure()

# define subplots and their positions in figure

plt1 = fig.add_subplot(221)   # 221 means 2 X 2 grid and refering to 1st subplot 
plt2 = fig.add_subplot(222)   # 222 means 2 X 2 grid and refering to 2nd subplot
plt3 = fig.add_subplot(223)   # 223 means 2 X 2 grid and refering to 3rd subplot
plt4 = fig.add_subplot(224)   # 224 means 2 X 2 grid and refering to 4th subplot

# plotting points on each subplot

x, y = create_plot('linear')
plt1.plot(x, y, color ='r')
plt1.set_title('$y_1 = x$')
 
x, y = create_plot('quadratic')
plt2.plot(x, y, color ='b')
plt2.set_title('$y_2 = x^2$')
 
x, y = create_plot('cubic')
plt3.plot(x, y, color ='g')
plt3.set_title('$y_3 = x^3$')
 
x, y = create_plot('quartic')
plt4.plot(x, y, color ='k')
plt4.set_title('$y_4 = x^4$')

# adjusting space between subplots

fig.subplots_adjust(hspace=.5,wspace=0.5)
plt.show()
plt.close()
```

### Plotting subplots using method 2.

```python
# function to generate coordinates

def create_plot(ptype):
    # setting the x-axis vaues
    x = np.arange(0, 5, 0.01)
     
    # setting y-axis values
    if ptype == 'sin':
        # a sine wave
        y = np.sin(2*np.pi*x)
    elif ptype == 'exp':
        # negative exponential function
        y = np.exp(-x)
    elif ptype == 'hybrid':
        # a damped sine wave
        y = (np.sin(2*np.pi*x))*(np.exp(-x))
             
    return(x, y)
```

```python
# defining subplots and their positions

plt1 = plt.subplot2grid((11,1), (0,0), rowspan = 3, colspan = 1)
plt2 = plt.subplot2grid((11,1), (4,0), rowspan = 3, colspan = 1)
plt3 = plt.subplot2grid((11,1), (8,0), rowspan = 3, colspan = 1)

# plotting points on each subplot

x, y = create_plot('sin')
plt1.plot(x, y, label = 'sine wave', color ='b')
x, y = create_plot('exp')
plt2.plot(x, y, label = 'negative exponential', color = 'r')
x, y = create_plot('hybrid')
plt3.plot(x, y, label = 'damped sine wave', color = 'g')

# show legends of each subplot

plt1.legend()
plt2.legend()
plt3.legend()

plt.show()
plt.close()
```

subplot2grid is similar to pyplot.subplot but uses 0-based indexing and let subplot to occupy multiple cells.
The arguments of the subplot2grid method:

* argument 1 : geometry of the grid
* argument 2 : location of the subplot in the grid
* argument 3 : (rowspan) No. of rows covered by subplot.
* argument 4 : (colspan) No. of columns covered by subplot.

>After going through the above two methods, we can infer that one should use subplot() method when the plots are of uniform size where as subplot2grid() method should be preferred when we want more flexibility on position and sizes of our subplots.
