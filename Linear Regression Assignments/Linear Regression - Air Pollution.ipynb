# Linear Regression - Air Pollution

## Problem Statement:
### Predicting Pollution Levels using Machine Learning
>It is winter time in Delhi so Cody decided to go for a walk to the news stand and on reaching
was surprised to see the air quality index of Delhi in the newspaper . So he decided to collect
the air samples of different locations and then took these samples to his lab where he extracted
five features of the air he collected which can be used to predict the air quality index and
combined it with the air quality index given in the newspapers. You are provided with the data
collected by Cody and your job is to design a machine learning model which given the features
extracted by Cody can predict air quality.

### Data consist of three csv files :-
1. Train.csv
2. Test.csv

> **Train:-** The train data consists of five feature columns and one target column. All the feature
columns are numeric and target value is also numeric.

>**Test:-** Test data consist all the columns contained by train with the exception of target column
which u have to predict using the model trained on train data.

```python
# Importing necessary Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing Training Data

df_Train = pd.read_csv('./Air Pollution Data/Training Data/Train.csv')

print(df_Train.shape, type(df_Train), df_Train, sep='\n\n')
```

```python
# Training Data into NDarray

X1Train = df_Train['feature_1'].values.reshape((-1,1))
X2Train = df_Train['feature_2'].values.reshape((-1,1))
X3Train = df_Train['feature_3'].values.reshape((-1,1))
X4Train = df_Train['feature_4'].values.reshape((-1,1))
X5Train = df_Train['feature_5'].values.reshape((-1,1))

XTrain = df_Train[['feature_1', 'feature_2', 'feature_3', 'feature_4', 'feature_5']].values
YTrain = df_Train['target'].values.reshape((-1,1))

print(X1Train.shape, type(X1Train), X1Train, X2Train, X3Train, X4Train, X5Train, sep='\n\n')
print('\n\n\n', XTrain.shape, type(XTrain), XTrain, YTrain.shape, type(YTrain), YTrain, sep='\n\n')
```

```python
# Visualising the Training Data

plt.figure(figsize=(20,20))
plt.scatter(X1Train, YTrain, color='violet', label='feature-1')
plt.scatter(X2Train, YTrain, color='green', label='feature-2')
plt.scatter(X3Train, YTrain, color='orange', label='feature-3')
plt.scatter(X4Train, YTrain, color='grey', label='feature-4')
plt.scatter(X5Train, YTrain, color='yellow', label='feature-5')

plt.title('Actual Output Given in Training Dataset')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.show()
```

```python
# Importing Scikit Learn Library

from sklearn.linear_model import LinearRegression
```

```python
#Creating an Object of LinearRegression Class

AirPollution_model = LinearRegression()
```

```python
# Training the AirPollution_model by fitting the Training Data into it

AirPollution_model.fit(XTrain, YTrain)
```

```python
# Prediction of the Output(y) of the features(x1, x2, x3, x4, x5) in the Training Dataset
# according to AirPollution_model we have created and trained.

output = AirPollution_model.predict(XTrain)
print(output.shape, type(output), output, sep='\n\n')
```

```python
# Getting the parameters of the predicted Output(y)
# (i.e., theta[0] and theta[1] or we can say that c and (m1, m2, m3, m4, m5) in the eq
# of the line: y = m1x1 + m2x2 + m3x3 + m4x4 + m5x5 + c)

bias = AirPollution_model.intercept_  # Intercept of the line in the equation of line (c)
coeff = AirPollution_model.coef_      # Coeffecient of the variables in the equation of line (m1, m2, m3, m4, m5)

print(bias, coeff, sep='\n\n')
```

```python
# Now Lets check the Accuracy of our model ---> AirPollution_model

AirPollution_model.score(XTrain, YTrain)    # Here, We are getting 96.61 % accuracy
```

```python
# Now, Lets Visualise the result we get for Training Dataset

plt.figure(figsize=(10,10))
plt.scatter(X1Train, YTrain, color='violet', label='Training Data - feature-1')
plt.scatter(X2Train, YTrain, color='green', label='Training Data - feature-2')
plt.scatter(X3Train, YTrain, color='orange', label='Training Data - feature-3')
plt.scatter(X4Train, YTrain, color='grey', label='Training Data - feature-4')
plt.scatter(X5Train, YTrain, color='yellow', label='Training Data - feature-5')

plt.plot(XTrain, output, color='lightblue', label='Predictions')

plt.title('Actual Output VS Predictions')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Now Importing the Test Dataset with values of features into a DataFrame

df_Test = pd.read_csv('./Air Pollution Data/Testing Data/Test.csv')

print(df_Test.shape, type(df_Test), df_Test, sep='\n\n')
```

```python
# Getting the Features from the DataFrame and saving them into a 2D Array using numpy

X1Test = df_Test['feature_1'].values.reshape((-1,1))
X2Test = df_Test['feature_2'].values.reshape((-1,1))
X3Test = df_Test['feature_3'].values.reshape((-1,1))
X4Test = df_Test['feature_4'].values.reshape((-1,1))
X5Test = df_Test['feature_5'].values.reshape((-1,1))

XTest = df_Test[['feature_1', 'feature_2', 'feature_3', 'feature_4', 'feature_5']].values

print(XTest.shape, type(XTest), XTest, sep='\n\n')
```

```python
# Now, Prediction of the Outputs(y) of the features(x1, x2, x3, x4, x5) in the Testing Dataset
# according to AirPollution_model we have created and trained. already

output_Test = AirPollution_model.predict(XTest)

print(output_Test.shape, type(output_Test), output_Test, sep='\n\n')
```

```python
# Lets Visualise our Predictions corresponding to the Testing Dataset Features

plt.figure(figsize=(20,15))
plt.scatter(X1Test, output_Test, color='violet', label='feature-1')
plt.scatter(X2Test, output_Test, color='green', label='feature-2')
plt.scatter(X3Test, output_Test, color='orange', label='feature-3')
plt.scatter(X4Test, output_Test, color='grey', label='feature-4')
plt.scatter(X5Test, output_Test, color='yellow', label='feature-5')

plt.title('Predictions For Testing Data Features')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.show()

plt.close()

plt.figure(figsize=(20,15))
'''plt.plot(XTest, output_Test, color='blue', label='Predictions')'''
plt.plot(X1Test, output_Test, color='violet', label='feature-1')
plt.plot(X2Test, output_Test, color='green', label='feature-2')
plt.plot(X3Test, output_Test, color='orange', label='feature-3')
plt.plot(X4Test, output_Test, color='grey', label='feature-4')
plt.plot(X5Test, output_Test, color='yellow', label='feature-5')

plt.title('Predictions For Testing Data Features')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.show()
```

```python
# Now, saving the the output in a DataFrame

OUT = df_Test

OUT['Predicted target'] = output_Test

OUT
```
