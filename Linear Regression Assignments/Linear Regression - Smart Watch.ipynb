# Linear Regression - Smart Watch

Since you are a student, you are expected to work hard and get better results than others. To track your time and expected performance in the Machine Learning Challenge, You are asked to install walkatime on your device which is an efficient time tracking tool to track your daily coding activity.

In this challenge, you have given walkatime data of some past students and how they performed in the evaluation exam. Your task is to predict the score you will get given the amount of time you spend on coding daily.

> **Input-** You are given one feature corresponding to time noted by walkatime.

> **Output-** A scalar denoting the level of performance students achieved by devoting the given time.

```python
# Importing necessary Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing Training Data

dfx = pd.read_csv('./Smart Watch Data/Training Data/Linear_X_Train.csv')
dfy = pd.read_csv('./Smart Watch Data/Training Data/Linear_Y_Train.csv')

print(dfx.shape, type(dfx), dfx, dfy.shape, type(dfx), dfy, sep='\n\n')
```

```python
# Training Data into NDarray

XTrain = dfx.values
YTrain = dfy.values
```

```python
print(XTrain.shape, type(XTrain), XTrain, YTrain.shape, type(YTrain), YTrain, sep='\n\n')
```

```python
# Visualising the Training Data

plt.figure(figsize=(20,15))
plt.scatter(XTrain, YTrain, color='red', label='Training Data')
plt.title('Actual Output Given in Training Data')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
# Importing Scikit Learn Library

from sklearn.linear_model import LinearRegression
```

```python
#Creating an Object of LinearRegression Class

SmartWatch_model = LinearRegression()
```

```python
# Training the SmartWatch_model by fitting the Training Data into it

SmartWatch_model.fit(XTrain, YTrain)
```

```python
# Prediction of the Outputs(y) of the features(x) in the Training Dataset
# according to SmartWatch_model we have created and trained

output = SmartWatch_model.predict(XTrain)
print(output.shape, type(output), output, sep='\n\n')
```

```python
# Getting the parameters of the predicted Output(y)
# (i.e., theta[0] and theta[1] or we can say that c and m in the line eq: y = mx + c)

bias = SmartWatch_model.intercept_     # Intercept of the line in the equation of line (c)
coeff = SmartWatch_model.coef_         # Coeffecient of the variable in the equation of line (m)

print(bias, coeff, sep='\n\n')
```

```python
# Now Lets check the Accuracy of our model ---> SmartWatch_model

SmartWatch_model.score(XTrain, YTrain)    # Here, We are getting 97.1 % accuracy
```

```python
# Now, Lets Visualise the result for Training Dataset

plt.figure(figsize=(20,15))
plt.scatter(XTrain, YTrain, color='red', label='Training Data')
plt.plot(XTrain, output, color='blue', label='Predictions')

plt.title('Actual Output VS Predictions For Training Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.show()
```

```python
# Now Importing the Test Dataset with values of features into a DataFrame

dfx_test = pd.read_csv('./Smart Watch Data/Testing Data/Linear_X_Test.csv')

print(dfx_test.shape, type(dfx_test), dfx_test, sep='\n\n')
```

```python
# Converting the DataFrame to a 2D Array using numpy

XTest = dfx_test.values

print(XTest.shape, type(XTest), XTest, sep='\n\n')
```

```python
# Now, Prediction of the Outputs(y) of the features(x) in the Testing Dataset
# according to SmartWatch_model we have created and trained already

output_Test = SmartWatch_model.predict(XTest)

print(output_Test.shape, type(output_Test), output_Test, sep='\n\n')
```

```python
# Lets Visualise our Predictions corresponding to the Testing Dataset Features

plt.figure(figsize=(20,15))
plt.scatter(XTest, output_Test, color='green', label='Predictions')

plt.title('Predictions For Testing Data Feature')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.show()

plt.close()

plt.figure(figsize=(20,15))
plt.plot(XTest, output_Test, color='blue', label='Predictions')

plt.title('Predictions For Testing Data Feature')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.show()
```

```python
# Now, saving the the output in a DataFrame

OUT = pd.DataFrame()

OUT['X-Feature'] = XTest.reshape(-1,)
OUT['PREDICTED OUTPUT'] = output_Test

OUT
```

```python
# Saving the predicted output corresponding to the Testing Dataset Features in a csv file

OUT.to_csv('./Smart Watch Data/Output/Test_Output.csv', index=False, header=True)
```
